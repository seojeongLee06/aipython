```python
# 사칙연산 프로그램 (변수 + 연산자 + 출력문만 사용)

a = 10  # 첫 번째 숫자
b = 3   # 두 번째 숫자

print("a =", a)  # a 값 출력
print("b =", b)  # b 값 출력

print("a + b =", a + b)  # 덧셈
print("a - b =", a - b)  # 뺄셈
print("a * b =", a * b)  # 곱셈
print("a / b =", a / b)  # 나눗셈(결과는 실수로 나옴)
```

```python
# 사칙연산 프로그램 (입력 + 변수 + 연산자 + 출력문)

a = int(input("첫 번째 숫자 a를 입력하세요: "))
b = int(input("두 번째 숫자 b를 입력하세요: "))

print("a =", a)
print("b =", b)

print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
```
