```python
fruits=["apple","banana","cherry"]
```

```python
fruits.extend(["orange","strawberry","kiwi","grapes","mango","pineapple","pear","berry","lemon"])
```

```python
len(fruits)
```

```python
for a in fruits:
  print(a)
```

```python
upper_fruits=[]
for fruit in fruits:
  upper_fruits.append(fruit.upper())
print(upper_fruits)
```

```python
for i, fruit in enumerate(fruits, start=1):
  print(f"{i}.{fruit}")
```

```python
for i, fruit in enumerate(fruits):
  print(f"{i+1}.{fruit}")
```
