```python
# 카페 메뉴(원)
menu_price = {
    "아메리카노": 4500,
    "카페라떼": 5200,
    "바닐라라떼": 5800,
    "카라멜마키아또": 5900,
    "카푸치노": 5200,
    "콜드브루": 5600,
    "카페모카": 6000,
    "에스프레소": 4000,
    "자몽에이드": 6500,
    "레몬에이드": 6500
}

orders = []  # (메뉴명, 수량, 금액) 저장

while True:
    menu = input("메뉴를 입력하세요 (종료 입력 시 끝): ").strip()

    if menu == "종료":
        print("주문을 종료합니다.")
        break

    if menu not in menu_price:
        print("해당 메뉴는 없습니다. 메뉴명을 다시 확인해주세요.")
        continue

    # 수량 입력 (정상 입력될 때까지)
    while True:
        qty_str = input("개수를 입력하세요: ").strip()
        if qty_str.isdigit() and int(qty_str) > 0:
            qty = int(qty_str)
            break
        print("개수는 1 이상의 정수로 입력해주세요.")

    price = menu_price[menu] * qty
    orders.append((menu, qty, price))

    print(f"{menu} {qty}개 가격은 {price}원입니다.")

# 마지막 주문 결과 출력
print("\n===== 주문 결과 =====")
if not orders:
    print("주문 내역이 없습니다.")
else:
    grand_total = 0
    for menu, qty, price in orders:
        print(f"- {menu} / {qty}개 / {price}원")
        grand_total += price
    print("---------------------")
    print(f"총 결제금액: {grand_total}원")
```
