```python
import random
answer=(random.randint(1, 100))
```

```python
answer
```

```python
user=int(input("추측값을 입력하시오"))
```

```python
count=0
while user!=answer:
  if user<answer:
    print("더 큰 값을 입력하시오")
    count+=1
  else:
    print("더 작은 수를 입력하시오")
    count+=1
  user=int(input("추측값을 입력하시오"))
print("총시도횟수",count)
```
